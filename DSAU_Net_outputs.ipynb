```python
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_data(input_names, output_type, output_axis, second_out, legend_names, title, max_epochs = 50):
    
    colour1 = "#1E88E5"
    colour2 = "#FFC107"
    
    loss_0 = []
    loss_Nan_0 = []
    figure(figsize=(14, 6), dpi=80)
    for c in range(len(output_type)):
        for i in range(max_epochs):

            loss_0 = np.genfromtxt(input_names + "/" + second_out[c] + "/epoch_" + str(i) + "_" + output_type[c] + ".csv",
                                   delimiter=',')
            loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))

        plt.plot(range(len(loss_Nan_0)),loss_Nan_0)
        loss_0 = []
        loss_Nan_0 = []

    plt.title(title,fontsize=20)
    plt.legend(legend_names, fontsize=16)
    plt.xlabel("Epoch",fontsize=18)
    plt.ylabel(ylabel=output_axis[c],fontsize=18)
    plt.grid()
    plt.show()  
```

```python
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_data_nomean(input_names, output_type, output_axis, second_out, legend_names, title):
    epoch = 0
    colour1 = "#1E88E5"
    colour2 = "#FFC107"
    
    loss_0 = []
    loss_Nan_0 = []
    figure(figsize=(14, 6), dpi=80)

    loss_0 = np.genfromtxt(input_names + "/" + second_out[0] + "/epoch_" + str(epoch) + "_" + output_type[0] + ".csv",delimiter=',')
    loss_1 = np.mean(loss_0[:(len(loss_0)//16)*16].reshape(-1,16), axis=1)
    plt.plot(range(len(loss_1)),loss_1)
    loss_0 = []
    loss_Nan_0 = []

    plt.title(title,fontsize=20)
    plt.legend(legend_names, fontsize=16)
    plt.xlabel("Epoch",fontsize=18)
    plt.ylabel(ylabel=output_axis[0],fontsize=18)
    plt.grid()
    plt.show()  

colour1 = "#1E88E5"
colour2 = "#FFC107"

sns.set(font_scale = 1.35)
sns.set_style("white",{"grid.color": ".5", "grid.linestyle": ":"})

input_names = "Checkpoints/UNet_DCSAU_Net_CT/Segmentation/"

output_type = ["loss"]
output_axis = ["dice loss"]
second_out = ["Training_loss/"]
legend_names = ["Training"]

title = "Training loss Epoch 0 - Segmentation model"

plot_data_nomean(input_names, output_type, output_axis, second_out, legend_names, title)

output_type = ["dice_score"]
output_axis = ["Dice"]
second_out = ["Training_Dice/"]

title = "Training dice Epoch 0 - Segmentation model"

plot_data_nomean(input_names, output_type, output_axis, second_out, legend_names, title)

output_type = ["loss"]
output_axis = ["dice loss"]
second_out = ["Validation_loss/"]
legend_names = ["Validation"]

title = "Training loss Epoch 0 - Segmentation model"

plot_data_nomean(input_names, output_type, output_axis, second_out, legend_names, title)

output_type = ["dice_score"]
output_axis = ["Dice"]
second_out = ["Validation_Dice/"]

title = "Training dice Epoch 0 - Segmentation model"

plot_data_nomean(input_names, output_type, output_axis, second_out, legend_names, title)
```
