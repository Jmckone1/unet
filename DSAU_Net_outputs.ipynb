```python
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_data(input_names, output_type, output_axis, second_out, legend_names, title, max_epochs = 50):
    
    colour1 = "#1E88E5"
    colour2 = "#FFC107"
    
    loss_0 = []
    loss_Nan_0 = []
    figure(figsize=(14, 6), dpi=80)
    for c in range(len(output_type)):
        for i in range(max_epochs):

            loss_0 = np.genfromtxt(input_names + "/" + second_out[c] + "/epoch_" + str(i) + "_" + output_type[c] + ".csv",
                                   delimiter=',')
            loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))

        plt.plot(range(len(loss_Nan_0)),loss_Nan_0)
        loss_0 = []
        loss_Nan_0 = []

    plt.title(title,fontsize=20)
    plt.legend(legend_names, fontsize=16)
    plt.xlabel("Epoch",fontsize=18)
    plt.ylabel(ylabel=output_axis[c],fontsize=18)
    plt.grid()
    plt.show()  
    
    
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_data2(input_names, output_type, output_axis, second_out, legend_names, title, max_epochs = 50):
    
    x = 0
    colour1 = "#1E88E5"
    colour2 = "#FFC107"
    
    loss_0 = []
    loss_Nan_0 = []
    figure(figsize=(14, 6), dpi=80)
    for c in range(len(output_type)):
        for i in range(max_epochs):

            loss_0 = np.genfromtxt(input_names + "/" + second_out[c] + "/epoch_" + str(i) + "_" + output_type[c] + ".csv",
                                   delimiter=',')
            
            for i in range(len(loss_0)):
                x = x + loss_0[i]
                if i % 64 == 0:
                    loss_Nan_0 = np.append(loss_Nan_0, x / 64)
                    x = 0
                #loss_Nan_0 = np.append(loss_Nan_0, (loss_0[np.logical_not(np.isnan(loss_0))]))

        plt.plot(range(len(loss_Nan_0)),loss_Nan_0)
        loss_0 = []
        loss_Nan_0 = []

    plt.title(title,fontsize=20)
    plt.legend(legend_names, fontsize=16)
    plt.xlabel("Epoch",fontsize=18)
    plt.ylabel(ylabel=output_axis[c],fontsize=18)
    plt.grid()
    plt.show()  
```
