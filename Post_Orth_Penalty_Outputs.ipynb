```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import pandas as pd

input_names = ["Unet_H16_M9_O10A0","Unet_H16_M10_O10"]
legend_names = ["10 orth model 9", "10 orthmodel 10"]
output_type = ["validation_loss", "validation_jaccard_index","training_loss","jaccard_index"]
output_axis = ["MSE loss + penalty", "jaccard_index","MSE loss + penalty","jaccard_index"]
output_title = ["validation loss", "validation Jaccard index","training loss","training Jaccard index"]

second_out = ["Validation_loss","Validation_Jaccard","Training_loss","Training_Jaccard"]

loss_0 = []
loss_Nan_0 = []

for c in range(len(output_type)):
    ax = plt.subplot()
    for s in range(len(input_names)):
        for i in range(18):
            if s == 1:
                loss_0 = np.genfromtxt("Checkpoints_RANO/"+ input_names[s]+ "/" + second_out[c] + "/epoch_"+str(i)+output_type[c] + ".csv",delimiter=',')
                loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))
            else:
                loss_0 = np.genfromtxt("Checkpoints_RANO/"+input_names[s]+"/epoch_"+str(i)+output_type[c],delimiter=',')
                loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))

        ax.plot(range(len(loss_Nan_0)),loss_Nan_0)
        loss_0 = []
        loss_Nan_0 = []
    if c == 0:
        plt.ylim([100, 400])

    plt.title("Mean " + output_title[c]+ " per Epoch ",fontsize=14)
    ax.legend(legend_names,fontsize=11)
    plt.xlabel("Epoch",fontsize=13)
    plt.ylabel(ylabel=output_axis[c],fontsize=13)
    plt.show()    
    
for i in range(35,45):
    loss_0 = np.genfromtxt("Checkpoints_RANO/Unet_H16_M9_O10A0/epoch_"+str(i)+"validation_jaccard_index",delimiter=',')
    loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))
print(loss_Nan_0)
```
