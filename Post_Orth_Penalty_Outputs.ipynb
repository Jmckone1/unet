```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import pandas as pd

input_names = ["Unet_H16_M13_O1000_cosine_orth_v2_small_half_v2","Unet_H16_M13_O10_cosine_orth_v2_small_half_v2_lr2"]
legend_names = ["lr: 0.0001", "lr: 0.0002"]
output_type = ["validation_loss", "validation_jaccard_index","training_loss","jaccard_index","training_loss_mse","training_loss_cosine","validation_loss_mse","validation_loss_cosine"]
output_axis = ["MSE loss + penalty", "jaccard_index","MSE loss + penalty","jaccard_index","training_loss_mse","training_loss_cosine","validation_loss_mse","validation_loss_cosine"]
output_title = ["validation loss", "validation Jaccard index","training loss","training Jaccard index","training_loss_mse","training_loss_cosine","validation_loss_mse","validation_loss_cosine"]

second_out = ["Validation_loss","Validation_Jaccard","Training_loss","Training_Jaccard","Training_loss_mse","Training_loss_cosine","Validation_loss_mse","Validation_loss_cosine"]

loss_0 = []
loss_Nan_0 = []

for c in range(len(output_type)):
    ax = plt.subplot()
    for s in range(len(input_names)):
        for i in range(100):
            loss_0 = np.genfromtxt("Checkpoints_RANO/"+ input_names[s]+ "/" + second_out[c] + "/epoch_"+str(i)+output_type[c] + ".csv",delimiter=',')
            loss_Nan_0 = np.append(loss_Nan_0, np.mean(loss_0[np.logical_not(np.isnan(loss_0))]))

        ax.plot(range(len(loss_Nan_0)),loss_Nan_0)
        loss_0 = []
        loss_Nan_0 = []

    plt.title("Mean " + output_title[c]+ " per Epoch ",fontsize=14)
    ax.legend(legend_names,fontsize=11)
    plt.xlabel("Epoch",fontsize=13)
    plt.ylabel(ylabel=output_axis[c],fontsize=13)
    plt.show()    
    
# one thing that i can do is to do as andy suggested and include a single parameters file that can be used to 
```
