```python
from os import walk
import csv
import numpy as np

d=[]
f=[]

for (dir_path, dir_names, file_names) in walk("Predictions/Validation_pred/"):
    f.extend(file_names)

for i in range(len(f)):
    a = f[i][-10:-8]
    print("a:", a)
    if a != "ct" and a != "re":
        d.append(int(a))
print("d: ",d)

array = np.array(d)
np.save("Val_data",d)
```

```python
import seaborn as sns
sns.set_theme()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train = np.load("Train_data.npy")
val = np.load("Val_data.npy")
test = np.load("Test_data.npy")

train2 = pd.DataFrame(data=train, columns=range(1)).assign(Data="Train")
val2 = pd.DataFrame(data=val, columns=range(1)).assign(Data="Val")
test2 = pd.DataFrame(data=test, columns=range(1)).assign(Data="Test")

cdf = pd.concat([train2, val2, test2])
mdf = pd.melt(cdf, id_vars=['Data'])

ax = sns.boxplot(x="Data", y="value", data=mdf)  # RUN PLOT   
plt.show()
ax = sns.violinplot(x="Data", y="value", data=mdf)  # RUN PLOT   
plt.show()

print(np.mean(val))
print(np.mean(test))

#f, axes = plt.subplots(1, 3)
#sns.boxplot(data=train2,  orient='v' , ax=axes[0])
#sns.boxplot(data=val2,  orient='v' , ax=axes[1])
#sns.boxplot(data=test2,  orient='v' , ax=axes[2])
```
