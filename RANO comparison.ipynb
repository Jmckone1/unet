```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()


x0 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_0training_dice.csv", delimiter=',')
x1 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_1training_dice.csv", delimiter=',')
x2 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_2training_dice.csv", delimiter=',')

#x = x0
x = np.concatenate((x0,x1,x2))

print("One Epoch equals" , len(x0)/(50*16))

mean_output = []
each_number = []
for i in range(len(x)):
    each_number.append(x[i])
    if len(each_number) == (50*16):
        mean_output.append(np.mean(each_number))
        each_number = []
    if i == len(x):
        mean_output.append(np.mean(each_number))
        
print("Base")
print(list(np.around(np.array(mean_output[:20]),3)))
        
plt.plot(range(len(mean_output)),mean_output )

x0 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_0training_dice.csv", delimiter=',')
x1 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_1training_dice.csv", delimiter=',')
x2 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_2training_dice.csv", delimiter=',')

#x = x0
x = np.concatenate((x0,x1,x2))

mean_output = []
each_number = []
for i in range(len(x)):
    each_number.append(x[i])
    if len(each_number) == (50*16):
        mean_output.append(np.mean(each_number))
        each_number = []
    if i == len(x):
        mean_output.append(np.mean(each_number))
        
print("Unfrozen")
print(list(np.around(np.array(mean_output[:20]),3)))

plt.plot(range(len(mean_output)),mean_output )
plt.title("Mean Training dice score for every 50 batches over 3 epochs")
plt.xlabel("Values per 50 batches (800 volumes)")
plt.ylabel("Mean Dice Similarity Score")
plt.legend(["Base", "Unfrozen"])

plt.show()
        
########################################################################################################################    

x0 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_0training_loss.csv", delimiter=',')
x1 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_1training_loss.csv", delimiter=',')
x2 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_base/epoch_2training_loss.csv", delimiter=',')

#x = x0
x = np.concatenate((x0,x1,x2))

mean_output = []
each_number = []
for i in range(len(x)):
    each_number.append(x[i])
    if len(each_number) == 50:
        mean_output.append(np.mean(each_number))
        each_number = []
    if i == len(x):
        mean_output.append(np.mean(each_number))
        
plt.plot(range(len(mean_output)),mean_output )

x0 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_0training_loss.csv", delimiter=',')
x1 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_1training_loss.csv", delimiter=',')
x2 = np.genfromtxt("Checkpoints/Full_model_MK5_H16_O4A4_unfrozen/epoch_2training_loss.csv", delimiter=',')

#x = x0
x = np.concatenate((x0,x1,x2))

mean_output = []
each_number = []
for i in range(len(x)):
    each_number.append(x[i])
    if len(each_number) == 50:
        mean_output.append(np.mean(each_number))
        each_number = []
    if i == len(x):
        mean_output.append(np.mean(each_number))
        
plt.plot(range(len(mean_output)),mean_output )
plt.title("Mean Training Loss for every 50 batches over 3 epochs")
plt.xlabel("Values per 50 batches (800 volumes)")
plt.ylabel("Mean loss - BCEWithLogitsLoss")
plt.legend(["Base", "Unfrozen"])
plt.show()
```
